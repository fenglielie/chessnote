```python
from chessnote import ChessState, ChessBoardRenderer, ChessRecorder
```

## ChessState

```python
print(ChessState())
print(ChessState(rotate=True))
print(ChessState(empty=True))
```

```python
state = ChessState().move((7, 2), (4, 2))  # change

state.pop((1, 9))  # delete
state[(2, 7)] = "h"  # add

print(state)
```

```python
state2 = state.deepcopy().rotate_pieces()
print(state2)

state3 = state.deepcopy().swap_pieces()
print(state3)
```

## ChessBoardRenderer

```python
ChessBoardRenderer().draw(ChessState())
ChessBoardRenderer().draw(ChessState().move((7, 2), (4, 2)))
```

```python
ChessBoardRenderer().draw(
    ChessState().move((1, 0), (2, 2)),
    highlight_pieces=[(1, 0), (2, 2)],
    arrows=[((1, 0), (2, 2))],
)
```

```python
ChessBoardRenderer().rotate().draw(
    ChessState().move((1, 0), (2, 2)).rotate_pieces(),
)

ChessBoardRenderer(rotate_flag=True).draw(
    ChessState(rotate=True).move((7, 9), (6, 7)),
)
```

## ChessRecorder

```python
recorder = ChessRecorder(ChessState(empty=True))
print(recorder)
```

```python
recorder = ChessRecorder().exec("炮二平五，马8进7")
print(recorder)
```

```python
data = """
1. 炮二平五 炮8平5
2. 马二进三 马8进7
3. 车一进一 车9平8
4. 车一平六 车8进6
5. 车六进七 马2进1
6. 车九进一 炮2进7
7. 炮八进五 马7退8
8. 炮五进四 士6进5
9. 车九平六 将5平6
10. 前车进一 士5退4
11. 车六平四 炮5平6
12. 车四进六 将6平5
13. 炮八平五
"""  # 弃马十三招

ChessRecorder().exec(data).draw(filename="demo.png")
```
