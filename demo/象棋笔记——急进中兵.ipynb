# 象棋笔记——急进中兵

```python
from chessnote import ChessRecorder
```

## 准备 

急进中兵是中炮过河车对屏风马平炮对车的一种常见进攻形式。我们直接从中炮过河车对屏风马，红方进七兵的局面开始讨论。

```python
r = ChessRecorder().exec(
    """
炮二平五，马8进7
马二进三，车9平8
车一平二，卒7进1
车二进六，马2进3
兵七进一，
"""
)
r.draw()
```

此时，黑方有两类主流走法：

1. 跳盘河马
2. 平炮兑车：比较稳健，防守反击

```python
r.dryrun(["马7进6", "炮8平9"])
```

这里主要关注平炮兑车后的变化。

```python
r.exec("炮8平9，车二平三，炮9退1").draw()
```

黑方接下来希望平炮打车，红方有多种应对思路：

1. 炮八平六，后续演化为五六炮
2. 马八进七，后续演化为五九炮
3. **兵五进一**，后续演化为“急进中兵”

这里主要关注兵五进一后的变化。

急进中兵的优点是进攻迅速，三路马可以从中路跳出，进而威胁对方的3路马。缺点是红方左侧车马炮出动较慢，进攻后劲不足。

黑方此时希望连起士象拱卫中路，首先必须用炮赶走三路车。

```python
r.exec("兵五进一，炮9平7，车三平四").draw()
```

黑方此时有三种选择来保护中路：

- 平炮（窝心炮）
- 飞象
- 拱士

### 黑方窝心炮 😈

黑方的一种做法是炮7平5形成窝心炮，此时虽然可以抑制红方中路进攻，但是窝心炮其实难以发出。

黑方不可以轻易打掉红方中兵，这是一个陷阱，此时红方会跳马邀兑，中路攻势仍然强劲。

```python
r1 = r.derive().exec("炮7平5，马八进七，炮5进4，马三进五")
r1.draw()
```

如果黑方选择兑炮，显然红方中路攻势继续。

```python
r1.derive().exec("炮5进2，炮八平五").draw()
```

黑方可以拱中卒给炮生根，但是会导致卒林线暴露，中路也很脆弱，红方可以在七路线继续进攻。

```python
r1.derive().exec("卒5进1，车四平七，马3进5，兵七进一").draw()
```

### 黑方飞象 😈

如果黑方飞象，红方会进车捉炮。而且轻易连象也导致马容易受到攻击。

```python
r.derive().exec("象3进5，车四进二").draw()
```

### 黑方拱士 ⭐

相比之下，拱士是更好的选择，因为可以用连环炮保护7路炮。

```python
r.derive().exec("士4进5，车四进二，炮2退1").draw()
```

```python
r.exec("士4进5").draw()
```

红方继续进中兵，就形成了标准的急进中兵局面，这是目前的一种常见阵势，与之形式类似的体系还有“牛头滚”，也是从中路发起进攻。

```python
r.exec("兵五进一").draw()
```

在急进中兵的局面下，黑方不能补相，否则红兵躲开，乘势上马。

```python
r.derive().exec("象3进5，兵五平六，车1平4，马三进五").draw()
```

黑方的走法有两类：

- 进中卒（吃中兵）
- 进7路卒（送卒）

```python
r.dryrun(["卒7进1", "卒5进1"])
```

## 黑方进7路卒

我们首先讨论一下进7路卒（送卒）的局面演化，这是一种主流的对攻走法，核心是在红方的三路线反击。

```python
r2 = r.derive().exec("卒7进1")
r2.draw()
```

### 红方吃卒 😈

黑方希望红方把卒吃掉，这样可以对红方三路线进行有效牵制。

```python
r21 = r2.derive().exec("兵三进一，象3进5")
r21.draw()
```

红方三路马有根，黑方看起来没有有效的反击，但是实际已经限制了红方两个马的移动，延缓红方进攻速度。

#### 红方跳三路马

一旦红方跳三路马，黑方会炮打底相进行对攻。注意底相是保不住的，车必须躲开，否则黑方跳马打死车。

```python
r21.derive().exec(
    "马三进五，马7进8，车四平三，马8退9，车三平四，炮7进8，士四进五，车8进9"
).draw()
```

到此为止，黑方优势明显，前炮后车，可以打士打马。

#### 红方跳七路马

如果红方马八进七，则三路马脱根，黑方跳马反击，同时打车打马，红方必须回车保马。

```python
r21.derive().exec("马八进七，马7进8，车四退四，卒5进1").draw()
```

到此为止，红方中路攻势被完全瓦解，中兵丢失，过河车撤回防守。

### 红方不吃卒 ⭐

红方应对黑方进7路卒的正确做法是不理睬，继续跳马进攻中路。

```python
r2.exec("马三进五").draw()
```

然后黑方吃卒，红方进马。

```python
r2.exec("卒7进1，马五进六").draw()
```

开局到此基本结束，进入中局，黑棋有两种选择：躲马或者弃马。

#### 黑方弃马

先看弃马的变化，黑车深入，既可以阻止红方飞相，也可以转移到2路线压马。

```python
r22 = r2.derive().exec("车8进8")
```

如果红方接受弃马，则黑方会平卒吃相，展开攻击

```python
r22.derive().exec("马六进七，卒7平6").draw()
```

如果红方不接收，选择跳八路马，黑方可以飞象准备出肋车，继续弃马。

```python
r22.derive().exec("马八进七，象3进5").draw()
```

#### 黑方躲马

黑方也可以马3退5选择躲马，看似软弱，实则仍然是对攻局面。

此时，红方吃中卒保持中路攻势，黑方可以继续7路线跳马反击，并且因为马跳走了，可以把2路炮也平过来，仍然有攻势。

```python
r23 = r2.derive().exec("马3退4")
r23.exec("兵五进一，马7进8").draw()
```

红方可以选择平兵将军，黑方可以平炮反将。

```python
r23.derive().exec("兵五平六，炮2平5").draw()
```

红方也可以选择退车到士角，后期选择兑车消解黑方对右侧的进攻。

```python
r23.derive().exec(
    "车四退四，炮7进8，士四进五，马8进9，车四平二，车8进7，炮八平二"
).draw()
```

## 黑方进5路卒 ⭐

相比于传统的进7路卒，进5路卒是现代才发展出的一种应对方式，也是更推荐的一类走法。

虽然吃中兵会暴露卒林线，黑方需要退马，但是此时黑方也可以进行平炮对攻，而且马可以从中路跳出，并不是很糟糕。

```python
r3 = r.derive().exec("卒5进1")
r3.draw()
```

此时红方吃卒，黑方退马。

```python
r3.exec("车四平七，马3退4").draw()
```

如果红方继续中路进攻，黑方可以还架中炮。

```python
r3.exec("马三进五，炮2平5").draw()
```

这里红方需要应对中路的反击，常见的做法包括：

- 炮打中卒
- 跳马躲开


### 红方炮打中卒 😈

此时红方不能炮打中卒，这是一个陷阱，黑方可以先上马逼迫红马连环，然后迅速出车，到中路吃马炮捉双。

```python
r31 = r3.derive().exec("炮五进三")
r31.exec("马7进6, 马八进七").draw()
```

注意这里的顺序，必须先跳马，使得红马连环，阻止红方炮八平五加强中路攻势。

```python
r31.exec("车1平2").draw()
```

红方必须进炮封车，否则黑方会进车过河，意图吃连环马和炮。

```python
r31.exec("炮八进四").draw()
```

黑方可以先换掉一个马，红马仍然被困在中路，可以从另一侧继续进车，到中路捉马。

```python
r31.exec("马6进5，马七进五，车8进5").draw()
```

### 红方跳马

如果红方跳马，黑方可以选择兑炮，成功瓦解红方的中路攻势。

```python
r32 = r3.derive().exec("马五进六")
r32.exec("炮5进5, 相七进五").draw()
```

黑方可以把马才中路跳出，逼红方兑掉骑河马，此时黑方彻底瓦解了攻势，还保留了中卒。

```python
r32.exec("马4进5，马六进五，象7进5").draw()
```
